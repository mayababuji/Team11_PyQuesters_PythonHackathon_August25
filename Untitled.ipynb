```python
import pandas as pd
import glob
import os
import numpy as np
from ydata_profiling import ProfileReport
```

```python
# Path where the 25 CSVs are stored
path_to_patient_files = '/Users/maya/Downloads/PyQuesters_Python_Hackathon/HUPA-UC Diabetes Dataset'
# Read and concatenate all the patient CSVs
all_files = glob.glob(os.path.join(path_to_patient_files, "*.csv"))
list_df = []
for filename in all_files:
    df = pd.read_csv(filename,sep=';')
    # Patient_ID is the filename without ".csv"
    df['Patient_ID'] = os.path.splitext(os.path.basename(filename))[0]
    list_df.append(df)

patient_df = pd.concat(list_df, ignore_index=True)
non_cleaned_df = pd.DataFrame(patient_df)
non_cleaned_df.to_csv("non_cleaned_df_output.csv")

'''convert the time from type object to datetime.
errors='coerce' means
--> If a value can be converted to a datetime → convert it.
--> If a value cannot be converted (e.g., "abc", "2020-13-45") →
instead of raising an error, replace it with NaT (Not-a-Time, the datetime version of NaN).'''
patient_df['time'] = pd.to_datetime(patient_df['time'], errors='coerce')
#No Negative Values Expected for insulin dose administered 
patient_df.loc[patient_df['bolus_volume_delivered'] < 0, 'bolus_volume_delivered'] = 0
## Round heart_rate to nearest integer for Clinical/medical analysis 
patient_df['heart_rate'] = patient_df['heart_rate'].round(0).astype(int)
# changing to  realistic basal_rate for modeling/analysis.
patient_df['basal_rate_cleaned'] = (patient_df['basal_rate'] / 0.025).round() * 0.025

# clip tiny noise values
patient_df.loc[patient_df['basal_rate_cleaned'] < 0.025, 'basal_rate_cleaned'] = 0
#Round calories to 1 decimal place
patient_df['calories_cleaned'] = patient_df['calories'].round(1)
#Round glucose to 1 decimal place
patient_df['glucose_cleaned'] = patient_df['glucose'].round(1)


cleaned_df = pd.DataFrame(patient_df)
cleaned_df.to_csv("cleaned_df_output.csv")
```

```python
#descriptive analysis 
'''Q1. How many participants in the dataset have diabetes, and what is their percentage?
Reasoning:
Since understanding the prevalence of diabetes is key to this analysis, examining the proportion of 
participants with the condition helps contextualize the population.
'''
import pandas as pd

# Load dataframe
df = pd.read_csv('cleaned_df_patient_race.csv')

# Count unique patients by gender
gender_counts = df.groupby('Patient_ID')['Gender'].first().value_counts()
gender_percent = gender_counts / gender_counts.sum() * 100

print("Number of participants by gender:")
print(gender_counts)
print("\nPercentage of participants by gender:")
print(gender_percent.round(2))
# Print which gender is higher
if gender_counts['Female'] > gender_counts['Male']:
    print("\nFemale participants are higher in number.")
elif gender_counts['Male'] > gender_counts['Female']:
    print("\nMale participants are higher in number.")
else:
    print("\nNumber of male and female participants are equal.")
```

```python
'''Q2. What is the average glucose for females and males?
Reasoning:
Comparing glucose by gender can highlight physiological differences or help identify at-risk groups.'''
# Average glucose by gender
avg_glucose_gender = df.groupby('Gender')['glucose_cleaned'].mean()
print(avg_glucose_gender.round(2))
#print which gender has more average glucose
if avg_glucose_gender['Female']>avg_glucose_gender['Male']:
    print("\nFemales in this dataset may have slightly higher average glucose readings.")
elif avg_glucose_gender['Female']< avg_glucose_gender['Male']:
     print("\nMales in this dataset may have slightly higher average glucose readings.")
elif avg_glucose_gender['Female']== avg_glucose_gender['Male']:
    print("\nMales and Females in this dataset  have equal average glucose readings.")
```

```python
'''Q3. What is the average sleep quality and average sleep duration of participants?

Reasoning:
Sleep affects glucose metabolism and insulin sensitivity. Understanding baseline sleep metrics helps contextualize glucose trends.'''
# Average sleep metrics
sleep_avg = df.groupby('Patient_ID')[['Average Sleep Duration (hrs)', 'Sleep Quality (1-10)']].first().mean()
print(sleep_avg.round(2))

if sleep_avg['Average Sleep Duration (hrs)']<7:
    print("\n Participants are sleep-deprived, which may affect glucose control, insulin sensitivity, heart health, and overall well-being")
elif sleep_avg['Average Sleep Duration (hrs)']>7 and sleep_avg['Average Sleep Duration (hrs)']<9:
    print("\n Participants are getting adequate sleep")
if sleep_avg['Sleep Quality (1-10)']<6:
    print("\n Participants sleep quality is moderate. Poor sleep quality can contribute to higher glucose variability, fatigue, and worse metabolic outcomes.")
elif sleep_avg['Sleep Quality (1-10)']>6:
    print ("\n Particants sleep quality is good")
```

```python
'''Prescriptive Question
Should patients with consistently high glucose (>180 mg/dL) increase their insulin dose, or
would lifestyle modifications (higher steps & lower carb intake) be more effective for glucose control?
Reasoning

Markers chosen:

glucose (outcome variable)

bolus_volume_delivered (insulin dose)

steps (physical activity)

carb_input (diet factor)

Medical rationale:

High glucose (>180 mg/dL) = hyperglycemia risk.

Clinically, both insulin adjustment and lifestyle changes (exercise + diet) influence glycemic control.

Prescriptive analysis should tell which intervention has stronger impact.

Analysis Approach

Identify hyperglycemic patients (glucose >180).

Compare correlations:

Glucose vs. Insulin dose.

Glucose vs. Steps.

Glucose vs. Carb intake.

Recommend the most effective factor based on correlation & visualization.'''
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Filter hyperglycemic patients
hyper_df = df[df["glucose"] > 180]

# Correlation analysis
corr = hyper_df[["glucose", "bolus_volume_delivered", "steps", "carb_input"]].corr()
print("Correlation Matrix:\n", corr)

# Visualization
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation of Glucose with Lifestyle & Insulin Factors")
plt.show()

# Prescriptive Recommendation
if abs(corr.loc["glucose", "carb_input"]) > abs(corr.loc["glucose", "bolus_volume_delivered"]):
    print("Recommendation: Focus on reducing carbohydrate intake for better glucose control.")
elif abs(corr.loc["glucose", "steps"]) > abs(corr.loc["glucose", "bolus_volume_delivered"]):
    print("Recommendation: Increasing physical activity (steps) is more effective than adjusting insulin alone.")
else:
    print("Recommendation: Insulin dose adjustment plays the stronger role in glucose control.")
```

```python
'''Prescriptive
“For patients with high glucose variability, should increasing 
basal insulin rate or reducing carbohydrate intake be prioritized for stabilizing glucose levels?
Why these markers?

Glucose Std Dev (variability) → poor glycemic control risk.

Basal Rate → long-term insulin support.

Carb Input → dietary impact on variability.”'''
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1️⃣ Compute glucose variability per patient (standard deviation)
variability = df.groupby("Patient_ID")["glucose"].std().reset_index()
variability.columns = ["Patient_ID", "glucose_variability"]

# 2️⃣ Aggregate mean basal rate and carb input per patient
basal_carb = df.groupby("Patient_ID")[["basal_rate", "carb_input"]].mean().reset_index()

# 3️⃣ Merge variability with basal and carb
merge_df = pd.merge(variability, basal_carb, on="Patient_ID")

# 4️⃣ Compute correlations
corr = merge_df[["glucose_variability", "basal_rate", "carb_input"]].corr()
print("Correlation matrix:\n", corr, "\n")

# 5️⃣ Visualize
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Glucose Variability vs Basal Insulin & Carb Intake")
plt.show()

# 6️⃣ Generate prescriptive recommendation
glucose_basal_corr = corr.loc["glucose_variability", "basal_rate"]
glucose_carb_corr = corr.loc["glucose_variability", "carb_input"]

if abs(glucose_basal_corr) > abs(glucose_carb_corr):
    recommendation = (
        f"Glucose variability is more strongly linked to basal insulin rate (corr={glucose_basal_corr:.2f}) "
        "than carbohydrate intake. For patients with high glucose variability, adjusting basal insulin may help stabilize glucose."
    )
else:
    recommendation = (
        f"Glucose variability is more strongly linked to carbohydrate intake (corr={glucose_carb_corr:.2f}) "
        "than basal insulin rate. For patients with high glucose variability, reducing carbohydrate intake may help stabilize glucose."
    )

print("PRESCRIPTIVE ANALYSIS RECOMMENDATION:")
print(recommendation)


```

```python
'''Prescriptive Question 2

“Do patients with higher physical activity (steps) require lower bolus insulin doses to maintain glucose control?”

Why these markers?

Steps = activity, increases insulin sensitivity.

Bolus Dose = short-term insulin to cover meals.

Glucose = outcome check.'''
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1️⃣ Aggregate mean values per patient
activity_df = df.groupby("Patient_ID")[["steps", "bolus_volume_delivered", "glucose"]].mean().reset_index()

# 2️⃣ Compute correlations (exclude Patient_ID)
corr = activity_df[["steps", "bolus_volume_delivered", "glucose"]].corr()
print("Correlation Matrix:\n", corr, "\n")

# 3️⃣ Generate prescriptive recommendation based on correlation
steps_bolus_corr = corr.loc["steps", "bolus_volume_delivered"]

if steps_bolus_corr < 0:
    recommendation = (
        f"Steps and bolus insulin have a negative correlation ({steps_bolus_corr:.2f}).\n"
        "This means that patients with higher physical activity tend to require slightly lower bolus doses to maintain glucose control."
    )
elif steps_bolus_corr > 0:
    recommendation = (
        f"Steps and bolus insulin have a positive correlation ({steps_bolus_corr:.2f}).\n"
        "This means that higher activity is associated with higher bolus doses, which is unusual and may need further investigation."
    )
else:
    recommendation = "No clear relationship between steps and bolus insulin in this dataset."

print("PRESCRIPTIVE ANALYSIS RECOMMENDATION:")
print(recommendation)

# 4️⃣ Visualize Steps vs Bolus Dose
sns.scatterplot(x="steps", y="bolus_volume_delivered", data=activity_df)
plt.title("Average Steps vs Average Bolus Dose per Patient")
plt.xlabel("Average Steps per Day")
plt.ylabel("Average Bolus Dose (units)")
plt.show()


```

```python
'''Prescriptive Question 3

“Is sleep duration more strongly associated with glucose control than insulin adjustment?”

Why these markers?

Sleep Hours (from demographics) → affects insulin sensitivity.

Glucose → main outcome.

Bolus/Basal Doses → medication adjustment.'''


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#  Average values per patient
sleep_df = df.groupby("Patient_ID")[["glucose", "bolus_volume_delivered", "basal_rate", "Average Sleep Duration (hrs)"]].mean().reset_index()

#  Compute correlations
corr = sleep_df[["glucose", "Average Sleep Duration (hrs)", "bolus_volume_delivered", "basal_rate"]].corr()
print("Correlation matrix:\n", corr)

#  Check which factor is most associated with glucose
glucose_corr = corr["glucose"].drop("glucose")  # remove self-correlation
strongest_factor = glucose_corr.abs().idxmax()
strength = glucose_corr[strongest_factor]

# Prescriptive recommendation
if strongest_factor == "Average Sleep Duration (hrs)":
    recommendation = f"Sleep duration has the strongest link to glucose (correlation={strength:.2f}). " \
                     "Improving sleep may help control glucose levels better than adjusting insulin alone."
else:
    recommendation = f"{strongest_factor.replace('_',' ').title()} has the strongest link to glucose (correlation={strength:.2f}). " \
                     "Focus on insulin adjustment to manage glucose levels."

print("\nPRESCRIPTIVE ANALYSIS RECOMMENDATION:")
print(recommendation)

# Visualize correlations
sns.heatmap(corr, annot=True, cmap="viridis")
plt.title("Glucose vs Sleep vs Insulin Factors")
plt.show()

```

```python
'''Prescriptive Question Example

Question:
“Do patients with higher average heart rate tend to burn more calories?”

Why these markers?

Heart Rate: Indicates intensity of activity.

Calories: Outcome measure for energy expenditure.'''
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1️⃣ Aggregate patient-level averages
hr_calories_df = df.groupby("Patient_ID")[["heart_rate", "calories"]].mean().reset_index()

# 2️⃣ Compute correlation
corr = hr_calories_df[["heart_rate", "calories"]].corr()
print("Correlation Matrix:\n", corr, "\n")

# 3️⃣ Generate recommendation based on correlation
hr_cal_corr = corr.loc["heart_rate", "calories"]

if hr_cal_corr > 0.4:
    recommendation = f"Heart rate and calories have a strong positive correlation ({hr_cal_corr:.2f}). Higher heart rate is strongly associated with more calories burned."
elif hr_cal_corr > 0.2:
    recommendation = f"Heart rate and calories have a moderate positive correlation ({hr_cal_corr:.2f}). Higher heart rate is moderately associated with more calories burned."
else:
    recommendation = f"Heart rate and calories have a weak correlation ({hr_cal_corr:.2f}). Heart rate alone may not strongly predict calories burned."

print("PRESCRIPTIVE ANALYSIS RECOMMENDATION:")
print(recommendation)

# 4️⃣ Visualize the relationship
sns.scatterplot(x="heart_rate", y="calories", data=hr_calories_df)
plt.title("Average Heart Rate vs Average Calories per Patient")
plt.xlabel("Average Heart Rate (BPM)")
plt.ylabel("Average Calories Burned")
plt.show()
```

```python
'''“Is sleep duration associated with average glucose levels?
Markers:

Average Sleep Duration (hrs): Sleep affects insulin sensitivity and hormone regulation. Poor or short sleep can lead to higher glucose.

Glucose: Main outcome measure for glycemic control.

Why these markers:

Sleep impacts how the body regulates blood sugar. By comparing sleep duration with glucose, we can see if longer or better sleep is associated with better glucose control.'''

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Aggregate patient-level averages
sleep_df = df.groupby("Patient_ID")[["Average Sleep Duration (hrs)", "glucose"]].mean().reset_index()

# Compute correlation
corr = sleep_df[["Average Sleep Duration (hrs)", "glucose"]].corr()
sleep_gl_corr = corr.loc["Average Sleep Duration (hrs)", "glucose"]
print("Correlation:", sleep_gl_corr)

# Recommendation
if sleep_gl_corr < -0.2:
    recommendation = f"Sleep and glucose have a negative correlation ({sleep_gl_corr:.2f}). Longer sleep is associated with lower glucose."
else:
    recommendation = f"Sleep and glucose have a weak correlation ({sleep_gl_corr:.2f}). Sleep alone may not strongly predict glucose."

print("Recommendation:", recommendation)

# Visualize
sns.scatterplot(x="Average Sleep Duration (hrs)", y="glucose", data=sleep_df)
plt.title("Sleep vs Glucose")
plt.show()
```

```python
'''Basal Insulin vs Glucose Variability

Question:
“For patients with high glucose variability, should basal insulin or carb reduction be prioritized?”
Markers:

Glucose Variability (Std Dev): Measures fluctuations in glucose; high variability is a risk for poor glycemic control.

Basal Insulin Rate: Long-term, continuous insulin support.

Carb Input: Dietary impact on glucose fluctuations.

Why these markers:

Patients with high glucose variability may benefit from either adjusting basal insulin or reducing carb intake. Comparing both helps prioritize interventions.'''
variability = df.groupby("Patient_ID")["glucose"].std().reset_index()
variability.columns = ["Patient_ID", "glucose_variability"]
basal_carb = df.groupby("Patient_ID")[["basal_rate","carb_input"]].mean().reset_index()
merge_df = pd.merge(variability, basal_carb, on="Patient_ID")

corr = merge_df[["glucose_variability","basal_rate","carb_input"]].corr()
glucose_basal_corr = corr.loc["glucose_variability","basal_rate"]
glucose_carb_corr = corr.loc["glucose_variability","carb_input"]

if abs(glucose_basal_corr) > abs(glucose_carb_corr):
    recommendation = f"Glucose variability is more associated with basal insulin (corr={glucose_basal_corr:.2f}). Adjust basal insulin to stabilize glucose."
else:
    recommendation = f"Glucose variability is more associated with carbohydrate intake (corr={glucose_carb_corr:.2f}). Reduce carb intake to stabilize glucose."

print("Recommendation:", recommendation)
```

```python
'''Prescriptive Question Example

Question:
“Do patients who have higher glucose levels tend to consume more carbohydrates later in the day?”

Why these markers:

Glucose: Outcome measure of glycemic control.

Carb Intake: Dietary factor that can raise blood sugar.

Rationale: Understanding whether high glucose levels are associated with higher carb intake helps determine if dietary adjustments may improve glucose control.'''
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1️⃣ Aggregate mean glucose and carb intake per patient
glucose_carb_df = df.groupby("Patient_ID")[["glucose", "carb_input"]].mean().reset_index()

# 2️⃣ Compute correlation
corr = glucose_carb_df[["glucose", "carb_input"]].corr()
glucose_carb_corr = corr.loc["glucose", "carb_input"]
print("Correlation Matrix:\n", corr, "\n")

# 3️⃣ Generate prescriptive recommendation based on correlation
if glucose_carb_corr > 0.4:
    recommendation = f"Glucose and carb intake have a strong positive correlation ({glucose_carb_corr:.2f}). Reducing carb intake may help control glucose."
elif glucose_carb_corr > 0.2:
    recommendation = f"Glucose and carb intake have a moderate positive correlation ({glucose_carb_corr:.2f}). Moderate carb adjustments may help control glucose."
else:
    recommendation = f"Glucose and carb intake have a weak correlation ({glucose_carb_corr:.2f}). Carb intake alone may not strongly predict glucose."

print("PRESCRIPTIVE ANALYSIS RECOMMENDATION:")
print(recommendation)

# 4️⃣ Visualize the relationship
sns.scatterplot(x="carb_input", y="glucose", data=glucose_carb_df)
plt.title("Average Carb Intake vs Average Glucose per Patient")
plt.xlabel("Average Carb Intake (grams)")
plt.ylabel("Average Glucose (mg/dL)")
plt.show()
```

```python
'''Steps vs Glucose Variability

Question:
“Do more active patients have lower glucose variability?”

Why these markers:

Steps: Indicator of daily physical activity.

Glucose Variability (Std Dev): Measures fluctuations in blood sugar.

Rationale: Physical activity improves insulin sensitivity, potentially reducing swings in glucose.'''
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1️⃣ Compute glucose variability (standard deviation) per patient
glucose_var = df.groupby("Patient_ID")["glucose"].std().reset_index()
glucose_var.columns = ["Patient_ID", "glucose_variability"]

# 2️⃣ Compute average steps per patient
steps_avg = df.groupby("Patient_ID")[["steps"]].mean().reset_index()

# 3️⃣ Merge the two
activity_var_df = pd.merge(glucose_var, steps_avg, on="Patient_ID")

# 4️⃣ Compute correlation
corr = activity_var_df[["steps", "glucose_variability"]].corr()
steps_var_corr = corr.loc["steps", "glucose_variability"]
print("Correlation between steps and glucose variability:", steps_var_corr)

# 5️⃣ Visualize the relationship
sns.scatterplot(x="steps", y="glucose_variability", data=activity_var_df)
plt.title("Steps vs Glucose Variability")
plt.xlabel("Average Steps per Day")
plt.ylabel("Glucose Variability (Std Dev)")
plt.show()

# 6️⃣ Generate prescriptive recommendation
if steps_var_corr < -0.4:
    recommendation = f"Strong negative correlation ({steps_var_corr:.2f}). More activity strongly reduces glucose variability."
elif steps_var_corr < -0.2:
    recommendation = f"Moderate negative correlation ({steps_var_corr:.2f}). Higher activity moderately reduces glucose variability."
else:
    recommendation = f"Weak correlation ({steps_var_corr:.2f}). Activity alone may not strongly reduce glucose variability."

print("PRESCRIPTIVE ANALYSIS RECOMMENDATION:")
print(recommendation)

```
